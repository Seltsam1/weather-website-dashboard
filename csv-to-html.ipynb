```python
# Dependencies

import pandas as pd
```

```python
# Read csv into a a dataframe

weather_data = pd.read_csv('./Resources/cities.csv')

weather_data
```

```python
# Reorder columns
weather_data = weather_data[["city_id", "City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]]
```

```python
# Convert dataframe to html

html_data = weather_data.to_html

html_data
```

```python
# Save html as file in Resources folder

weather_data.to_html(open('./Resources/cities.html', 'w'))
```
